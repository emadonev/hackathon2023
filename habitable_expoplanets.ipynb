## Animating planet circling around star

```python
import matplotlib.pyplot as plt
import numpy as np
# Initialize figure and axis
fig, ax = plt.subplots()

# Set axis limits
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

# Draw the central circle
central_circle = plt.Circle((0, 0), 1, fill=False, color='blue', linewidth=2)
ax.add_artist(central_circle)

# Parameters for the orbiting circle
radius_orbit = 5
angle = np.linspace(0, 2*np.pi, 100)

# Calculate the coordinates for the orbiting circle
x_orbit = radius_orbit * np.cos(angle)
y_orbit = radius_orbit * np.sin(angle)

# Plot the orbiting circle at different angles
for a in angle:
    x = radius_orbit * np.cos(a)
    y = radius_orbit * np.sin(a)
    orbiting_circle = plt.Circle((x, y), 0.5, fill=False, color='red', linewidth=2)
    ax.add_artist(orbiting_circle)
    plt.pause(0.1)  # Pause for a short period
    orbiting_circle.remove()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
def plot_color_gradient(cmap_name):
    gradient = np.linspace(0, 1, 256)
    gradient = np.vstack((gradient, gradient))
    fig, ax = plt.subplots(figsize=(6.4,.5))
    ax.imshow(gradient, aspect='auto', cmap=plt.get_cmap(cmap_name))
    
plot_color_gradient('Reds')
```

```python
colors=["white","blue","black"]
nodes = [0.0, 0.5, 1.0]
blue_cmap = LinearSegmentedColormap.from_list("blue_cmap", list(zip(nodes, colors)))
plot_color_gradient(blue_cmap)
```

```python
import numpy as np
import re
import pandas as pd

index = 0

with open('podaci.txt') as f:
    lines = f.readlines()

data = {}

for line in lines:
    if ":" not in line:
        continue  # Skip lines without a colon
    key, values_str = line.split(":")
    values_str = re.sub('[^0-9. ,]', '', values_str)
    values_list = values_str.split(', ')
    values_list = [float(x) for x in values_list]
    data[key.strip()] = values_list

specific_planet = []
for dic in data:
    specific_planet.append(data[dic][index])

print(specific_planet)
```
